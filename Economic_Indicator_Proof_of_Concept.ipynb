**Economic Indicator - Proof of Concept**

This notebook takes a dataset about grocery stores from data.gov, manipulates it using Pandas, and then feeds it to Tableau to visualize the density of groceries stores per zip code in New York.




Source: https://catalog.data.gov/dataset/retail-food-stores/resource/fa46d745-a3ef-41ea-a97d-e4f4827543e3

```python
# Import Pandas for data manipulation and analysis
import pandas as pd
```

```python
# Creates dataframe from the .csv Retail_Food_Stores.csv and prints the top 10 entries
ny_food_stores_df = pd.read_csv('Retail_Food_Stores.csv')
ny_food_stores_df.head(10)
```

```python
# Pulls and prints Zip Codes column from ny_food_stores_df
ny_food_stores_zips = ny_food_stores_df['Zip Code']
ny_food_stores_zips.head(5)
```

```python
# Counts and prints the number of occurences per uniqe zip code in ny_food_stores_zips
zips_count = (ny_food_stores_zips.value_counts())
zips_count.head(5)
```

```python
# Converts ny_food_stores_zips into a .csv
zips_count = pd.DataFrame(zips_count)
zips_count.to_csv('zips_count.csv' ,  index = True)
```

```python
# Converts zips_count.csv into a dataframe for heading manipulation, changes heading, converts back to csv and prints new datafram
zips_final = pd.read_csv("zips_count.csv")
zips_final.columns = ['Zip Code', 'Frequency']
zips_final.to_csv('zips_final.csv' ,  index = True)
zips_final.head(10)
```
